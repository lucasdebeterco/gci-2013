#Aula 10

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.ensemble import RandomForestClassifier
```

```python
estados = pd.read_csv('brasil_estados.csv')
estados
```

```python
## preparação dos dados
variaveis = ['area', 'populacao', 'densidade',
             'matricula', 'idh', 'receitas',
             'despesas', 'rendimento', 'veiculos']
```

```python
data = estados[variaveis].copy()
data['regiao'] = estados['regiao_nome']
data
```

```python
## enriquecimento dos dados
X = data.drop(columns='regiao') ## variáveis do modelo
Y = data['regiao'] ## variável dependente
```

```python
## modelo de inteligência --> classificação --> Decision Tree
decision_tree = DecisionTreeClassifier(
    random_state=0, max_depth=4
)
decision_tree = decision_tree.fit(X, Y)
```

```python
## visualização do resultado
print(export_text(decision_tree, feature_names=variaveis))

## Modelo Random Forest
from sklearn.preprocessing import StandardScaler
```

```python
## Enriquecimento Treino/Teste
X_Train = data.drop(columns='regiao', axis=1)
X_Test = data.drop(columns='regiao', axis=1)
y_Train = data['regiao']
y_Test = data['regiao']
## valores padronizados
sc_x = StandardScaler()
X_Train = sc_x.fit_transform(X_Train)
X_Test = sc_x.fit_transform(X_Test)
```

```python
forest = RandomForestClassifier(random_state=0)
## Ajuste/Treina
forest.fit(X_Train, y_Train)
## Aplicar
predicao = forest.predict((X_Test))
predicao

data['regiao_previsao'] = predicao

data
```

```python
importancia = forest.feature_importances_
importancia = pd.DataFrame(importancia)
importancia.columns = ['importancia']
importancia['variavel'] = variaveis
importancia.sort_values(by='importancia', ascending=False)
```

#Aula 11

```python
import pandas as pd
import xlrd
import numpy as np
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

estados = pd.read_csv('brasil_estados.csv')
estados
```

```python
variaveis = [
    'area',
    'populacao',
    'densidade',
    'matricula',
    'idh',
    'receitas',
    'despesas',
    'rendimento',
    'veiculos'
]
```

```python
#Clusterização:

data = estados[variaveis].copy()
kmeans = KMeans(n_clusters=5, random_state=0).fit(data)
datac = estados[['regiao_nome','uf']].copy()
datac = datac.merge(data, left_index=True, right_index=True)
datac['cluster'] = kmeans.labels_
datas = datac.sort_values(by='cluster')
datas

```

```python
centroides = kmeans.cluster_centers_
centroides = pd.DataFrame(
    centroides,
    columns = variaveis
)
centroides = centroides.reset_index()
centroides = centroides.rename(columns = {'index': 'cluster'})
centroides
```

```python
#PCA:

data = estados[variaveis].copy()
pca = PCA(n_components=2).fit(data)
datax = pca.transform(data)
datap = estados[['regiao_nome','uf']].copy()
datap = datap.merge(data, left_index=True, right_index=True)
datap['escalaX'] = datax[:, 0]
datap['escalaY'] = datax[:, 1]
datap
```

#Aula 13

```python
import pandas as pd
import xlrd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

estados = pd.read_csv('brasil_estados.csv')
estados
```

```python
variaveis = [
    'area',
    'populacao',
    'densidade',
    'matricula',
    'idh',
    'receitas',
    'despesas',
    'rendimento',
    'veiculos'
]
```

```python
data = estados[variaveis].copy()
data['regiao'] = estados['regiao_nome']
X_Train = data.drop(columns=['regiao'], axis=1)
X_Test = data.drop(columns=['regiao'], axis=1)
y_Train = data['regiao']
y_Test = data['regiao']
sc_x = StandardScaler()
X_Train = sc_x.fit_transform(X_Train)
X_Test = sc_x.fit_transform(X_Test)
logreg = LogisticRegression(solver="lbfgs", max_iter=500)
logreg.fit(X_Train, y_Train)
```

```python
pred_logreg = logreg.predict(X_Test)
pred_proba = logreg.predict_proba(X_Test)
```

```python
datar = estados[['regiao_nome','uf']].copy()
datar = datar.merge(data, left_index=True, right_index=True)
datar["regiao_pred"] = pred_logreg
datar
```

```python
lista_proba = pred_proba.tolist()
lista_proba = pd.DataFrame(
    lista_proba,
    columns = [
        'CentroOeste', 'Nordeste', 'Norte', 'Sudeste', 'Sul'
    ]
)
lista_proba
```

```python
datar = pd.merge(datar, lista_proba, left_index=True, right_index=True)
datar
```

```python
from sklearn.preprocessing import StandardScaler
from pandas.api.types import is_numeric_dtype
from sklearn.neighbors import NearestNeighbors
from pyod.models.knn import KNN
```

```python
estados = pd.read_csv('brasil_estados.csv')
estados
```

```python
variaveis = [
    'area',
    'populacao',
    'densidade',
    'matricula',
    'idh',
    'receitas',
    'despesas',
    'rendimento',
    'veiculos'
]
```

```python
data = estados[variaveis].copy()
for estado in estados['uf']:
    print(estado)
    datax = estados[estados['uf'] == estado][variaveis]
    neighbors_alg = NearestNeighbors(n_neighbors=min(5, len(data))).fit(data)
    similar = neighbors_alg.kneighbors(datax, return_distance=False)[0]
    data_similar = estados.iloc[similar][variaveis].reset_index()
    df_uf = estados.iloc[similar]['uf'].reset_index()
    df_uf = df_uf.merge(data_similar, on='index', how='left')
    df_uf = df_uf.drop(columns=['index'])
    print(df_uf)
```

```python
sc_x = StandardScaler()
datax = sc_x.fit_transform(data)
clf = KNN().fit(data)
outliers = clf.predict(data)
outliers
```

```python
dataod = estados.copy()
dataod['outlier'] = outliers
dataod

dataod[dataod['outlier'] == 1]
```
