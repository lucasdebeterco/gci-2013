```python
import pandas as pd
import datetime as dt

from sklearn.linear_model import LinearRegression
from prophet import Prophet

import seaborn as sns

from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

import logging, sys
logging.disable(sys.maxsize)

gs = pd.read_excel('../GS.xlsx')
gs

gs['ano'] = gs['Order Date'].dt.year
gs['mes'] = gs['Order Date'].dt.month
gs['periodo'] = ((gs['ano'] -  gs['ano'].min()) * 12) + gs['mes']
gs['data'] = gs['Order Date'].dt.strftime("%Y-%m-01")
gs['data']= pd.to_datetime(gs['data'])
gs

## Regressão Linear
gs_cat = gs.groupby(['Sub-Category','periodo'])['Sales'].sum().reset_index().copy()
gs_cat

for cat in gs_cat['Sub-Category'].unique():
    print(cat)
    df_cat = gs_cat[gs_cat['Sub-Category'] == cat][['periodo','Sales']]
    sns.lineplot(data=df_cat, x="periodo", y="Sales")
    plt.show()
    X = df_cat['periodo'].values.reshape(-1, 1)
    y = df_cat['Sales'].values.reshape(-1, 1)
    regressor = LinearRegression()
    regressor.fit(X, y)
    A = regressor.intercept_[0]
    B = regressor.coef_[0][0]
    x = df_cat['periodo'].max() + 1
    Y = A + (B * x)
    print(f'categoria {cat} intercept {A} coefficient {B} projecao {Y}')

gs_cat = gs.groupby(['Sub-Category','data'])['Sales'].sum().reset_index().copy()
gs_cat

gs_cat.info()

for cat in gs_cat['Sub-Category'].unique():
    print(cat)
    df_plot = gs_cat[gs_cat['Sub-Category'] == cat]
    df_plot = df_plot[['data','Sales']]
    df_plot = df_plot.rename(
       columns={
            'data': 'ds',
            'Sales': 'y'
        }
    )
    m = Prophet().fit(df_plot)
    future = m.make_future_dataframe(periods=12, freq='MS')
    forecast = m.predict(future)
    forecast = forecast.merge(df_plot, on='ds', how='left')
    print(forecast[['ds', 'y', 'yhat', 'yhat_lower', 'yhat_upper']])
    fig1 = m.plot(forecast)
    plt.show()
```
