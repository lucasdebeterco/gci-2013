```python
import pandas as pd
import xlrd
import numpy as np
from sklearn.preprocessing import StandardScaler
from pandas.api.types import is_numeric_dtype
from sklearn.neighbors import NearestNeighbors
from pyod.models.knn import KNN

estados = pd.read_csv('../brasil_estados.csv')
estados

variaveis = [
    'area',
    'populacao',
    'densidade',
    'matricula',
    'idh',
    'receitas',
    'despesas',
    'rendimento',
    'veiculos'
]

data = estados[variaveis].copy()
for estado in estados['uf']:
    print(estado)
    datax = estados[estados['uf'] == estado][variaveis]
    neighbors_alg = NearestNeighbors(n_neighbors=min(5, len(data))).fit(data)
    similar = neighbors_alg.kneighbors(datax, return_distance=False)[0]
    data_similar = estados.iloc[similar][variaveis].reset_index()
    df_uf = estados.iloc[similar]['uf'].reset_index()
    df_uf = df_uf.merge(data_similar, on='index', how='left')
    df_uf = df_uf.drop(columns=['index'])
    print(df_uf)
```

```python
sc_x = StandardScaler()
datax = sc_x.fit_transform(data)
clf = KNN().fit(data)
outliers = clf.predict(data)
outliers

dataod = estados.copy()
dataod['outlier'] = outliers
dataod

dataod[dataod['outlier'] == 1]
```
